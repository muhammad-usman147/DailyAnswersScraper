```python
#for answers
query  = {}
answer = []
count = 0
bodyimage = []
desc = []
fi = []
pds = []
pu = []
title = []
for i in range(len(data)):
    x = data.iloc[i,:]
    if pd.isnull(x.answer) == False:
        answer.append(x.answer)
        bodyimage.append(x.body_image_url)
        desc.append(x.description)
        fi.append(x.featured_image_url)
        pds.append(x.post_date)
        pu.append(x.post_url)
        title.append(x.title)
        count+=1
print("Total are {}".format(count))
```
